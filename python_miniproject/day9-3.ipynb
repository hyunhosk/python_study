```python
import pandas as pd

df = pd.read_csv('../data/dirtydata.csv', encoding="utf-8")

df
```

```python
pd.get_dummies(df['Duration'])
```

# binning

```python
# labels에 원래는 0,1같이 숫자 넣어야함
bins = [df['Maxpulse'].min(), 120, df['Maxpulse'].max()]
df['MaxP'] = pd.cut(x = df['Maxpulse'], bins = bins, labels = ['Low', 'High'])
df
```

```python
bins = [df['Calories'].min(), 320, df['Calories'].max()]
df['avCal'] = pd.cut(x = df['Calories'], bins = bins, labels = [1, 2])
df
```
