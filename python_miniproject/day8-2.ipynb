```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.rc('font', family='Malgun Gothic')

plt.plot(x, y, '--y')
plt.title('한글테스트')
plt.xlabel('한글')
plt.ylabel('테스트')

plt.show()
```
