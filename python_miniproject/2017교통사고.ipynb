```python
import pandas as pd
data = pd.read_csv('2017교통사고현황.csv')

data
```

```python
df_aci = pd.DataFrame({'발생일' : data['발생일'],
                         '사고건수' : data['사고건수']})
df_aci
```

```python
df_aci.loc[0]['발생일']
```

```python
df = pd.DataFrame()
day = '0'
for i in df_aci.loc():
    tot_aci = 1
    if i['발생일'] == day:
        tot_aci += 1
    else:
        df['발생일'] = day
        df['사고건수'] = tot_aci
        a = i['발생일']
df
```

다르면 날자(a)  / 사고몇번(tot_aci) 를 입력

```python
df = pd.DataFrame({'발생일' : a,
                   '사고건수' : tot_aci})
```
