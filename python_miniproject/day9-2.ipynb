# dataframe

```python
import pandas as pd
```

```python
fruit = pd.DataFrame({'Num' : [123, 456, 789, 1011, 1112],
                      'Fruit' : ['apple', 'banana', 'cherry', 'lemon', 'peach']})

grade = pd.DataFrame({'Num' : [123, 789, 1314],
                      'Grade' : ['A', 'B', 'C']})

price = pd.DataFrame({'No' : [123, 789, 1314],
                      'Grade' : [100, 200, 500]})

print(fruit)
print(grade)
print(price)
```

정보 합치기

```python
df = pd.concat([fruit, grade, price])
df
```

# merge ( 교집합 찾기)

```python
# 아무 조건없이 merge 하면 겹치는 부분만 출력 (inner)
df = pd.merge(fruit, grade, on = 'Num')
df
```

```python
df = pd.merge(fruit, grade, on='Num', how = 'left')
df
```

```python
df = pd.merge(fruit, grade, on='Num', how = 'right')
df
```

```python
df = pd.merge(fruit, grade, on='Num', how = 'outer')
df
df.isnull().sum()
```

```python
# join - 열 이름 다를 때 합치는 법. 왼쪽 프레임(fruit)의 Num, 오른쪽 프레임(price)의 No
df = pd.merge(fruit, price, left_on = 'Num', right_on = 'No', how = 'left')
df.isnull().sum()
```
